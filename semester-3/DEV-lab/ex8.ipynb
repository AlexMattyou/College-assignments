## EDA on e-mail dataset:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("./dataset/email.csv")
df.head()
```

```python
df.describe(include='all')
df.info()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Labels', data=df,)
plt.xticks(rotation=90)
plt.show()
```

```python
df['From'].value_counts().plot(kind='bar', title='From')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(df['From'],df['Date'])
plt.xticks(rotation=90)
plt.show()
```

```python
df['From'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
