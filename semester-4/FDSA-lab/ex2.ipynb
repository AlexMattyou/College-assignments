```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Line plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Line Plot', color='b', marker='o')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.legend()
plt.grid(True)
plt.show()

# Bar plot
plt.figure(figsize=(10, 5))
plt.bar(x, y, color='g', alpha=0.7)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Bar Plot')
plt.show()

# Scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='r', marker='x')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Scatter Plot')
plt.show()
```
