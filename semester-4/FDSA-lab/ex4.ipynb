```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
data = {
    'Age': [24, 27, 22, 32, 29, 24, 22, 27, 30, 27],
    'Salary': [70000, 80000, 50000, 90000, 60000, 70000, 50000, 80000, 75000, 65000]
}

df = pd.DataFrame(data)

# Normal curve for Age
mean_age = df['Age'].mean()
std_age = df['Age'].std()
age_range = np.linspace(df['Age'].min(), df['Age'].max(), 100)
normal_curve_age = norm.pdf(age_range, mean_age, std_age)

plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=10, density=True, alpha=0.6, color='g', edgecolor='black')
plt.plot(age_range, normal_curve_age, 'r')
plt.title('Normal Curve for Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# Scatter plot and correlation coefficient for Age vs Salary
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Salary'], alpha=0.7)
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

correlation_coef = df['Age'].corr(df['Salary'])
print("Correlation Coefficient between Age and Salary:", correlation_coef)
```
