```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack'],
    'Age': [24, 27, 22, 32, 29, 24, 22, 27, 30, 27],
    'Salary': [70000, 80000, 50000, 90000, 60000, 70000, 50000, 80000, 75000, 65000]
}

df = pd.DataFrame(data)

# Frequency distribution of Ages
age_freq = df['Age'].value_counts()
print("Frequency Distribution of Ages:\n", age_freq)

# Averages and Variability
mean_age = df['Age'].mean()
median_age = df['Age'].median()
std_age = df['Age'].std()
var_age = df['Age'].var()
print("\nMean Age:", mean_age)
print("Median Age:", median_age)
print("Standard Deviation of Age:", std_age)
print("Variance of Age:", var_age)

# Plotting frequency distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=range(20, 35), edgecolor='black', alpha=0.7)
plt.title('Frequency Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
